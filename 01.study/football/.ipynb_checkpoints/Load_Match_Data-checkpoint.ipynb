```python
def read_json_file(filename):
    with open(filename, 'rb') as json_file:
        return BytesIO(json_file.read()).getvalue().decode('unicode_escape')
```

```python
### 선수 만나이 계산을 위한 나이 계산기 함수 정의
def year_old(birth) :
    birth = datetime.strptime(birth, '%Y-%m-%d').date()
    today = datetime.now().date()
    
    age_day = today.day - birth.day
    age_month = today.month - birth.month
    age_year = today.year - birth.year
    
    #일 계산
    if age_day < 0 :
        age_month -= 1
        last_month_last_day = (today.replace(day=1)-timedelta(days=1)).day
        #지난 달의 말일 = 지난 달의 날수
        
        if last_month_last_day > birth.day : # 생일이 29일, 지난달 수가 30일인 경우
            age_day += last_month_last_day
        else : # 생일이 31일, 지난달 수가 30일인 경우
            age_day = today.day
            
    if age_month < 0 :
        age_year -= 1
        age_month += 12
    
    yr = str(age_year)+'yr'
    
    return yr
```

```python
competitions = [
     'England',
     'France',
     'Germany',
     'Italy',
    'Spain',
    'European Championship',
    'World Cup'
]

dfs_matches = []
for competition in competitions:
    competition_name = competition.replace(' ', '_')
    file_matches = f'matches_{competition_name}.json'
    json_matches = read_json_file(file_matches)
    df_matches = pd.read_json(json_matches)
    dfs_matches.append(df_matches)
df_matches = pd.concat(dfs_matches)
```

```python
df_matches.to_hdf('wyscout.h5', key='matches', mode='a')
```

```python
for competition in competitions:
    competition_name = competition.replace(' ', '_')
    file_events = f'events_{competition_name}.json'
    json_events = read_json_file(file_events)
    df_events = pd.read_json(json_events)
    df_events_matches = df_events.groupby('matchId', as_index=False)
    for match_id, df_events_match in df_events_matches:
        df_events_match.to_hdf('wyscout.h5', key=f'events/match_{match_id}', mode='a')
```
